```python
from google.colab import files
import io
import pandas as pd
import numpy as np
from pycaret.regression import RegressionExperiment
```

```python
uploaded = files.upload()
total_data = pd.read_csv(io.BytesIO(uploaded["winequality-red.csv"]))

train = total_data.sample(frac = 0.8)
test = total_data.drop(train.index).drop("quality", axis = 1)
```

```python
train.head()
```

```python
test.head()
```

```python
s = RegressionExperiment()
s.setup(data = train, target = "quality", use_gpu = True)
```

```python
s.compare_models()
```

```python
model = s.create_model("et")
```

```python
s.evaluate_model(model)
```

```python
train_holdout = s.predict_model(model)
```

```python
predictions = s.predict_model(model, data = train)
s.save_model(model, "best")

predictions
```
