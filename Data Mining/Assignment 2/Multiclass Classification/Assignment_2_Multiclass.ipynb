```python
from google.colab import files
import io
import pandas as pd
from pycaret.classification import ClassificationExperiment

training_file = files.upload()
training_data = pd.read_csv(io.BytesIO(training_file["train.csv"]))

testing_file = files.upload()
testing_data = pd.read_csv(io.BytesIO(testing_file["test.csv"]))

testing_data.head()
training_data.head()
```

```python
s = ClassificationExperiment()
s.setup(data = training_data, target = "Segmentation", session_id = 123, use_gpu = True)
s.compare_models()
```

```python
model = s.create_model("gbc")
```

```python
tuned = s.tune_model(model)
```

```python
training_holdout = s.predict_model(tuned)
```

```python
test_predictions = s.predict_model(tuned, data = testing_data)
test_predictions
```
