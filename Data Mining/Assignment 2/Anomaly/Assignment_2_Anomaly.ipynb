```python
from pycaret.anomaly import AnomalyExperiment
import pandas as pd
from google.colab import files
import io
```

```python
uploaded = files.upload()
data = pd.read_csv(io.BytesIO(uploaded["network.csv"]))
data.head()
```

```python
exp = AnomalyExperiment()
exp.setup(data = data, session_id = 112, use_gpu = True)
```

```python
iforest = exp.create_model("iforest")
result = exp.assign_model(iforest)
exp.evaluate_model(iforest)
```

```python
new_data = data.copy()
predictions = exp.predict_model(iforest, data = new_data)
predictions
```
