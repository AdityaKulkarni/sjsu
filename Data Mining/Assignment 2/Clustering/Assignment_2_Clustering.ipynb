```python
from pycaret.clustering import ClusteringExperiment
import pandas as pd
from google.colab import files
import io
```

```python
uploaded = files.upload()
data = pd.read_csv(io.BytesIO(uploaded["CC GENERAL.csv"]))
```

```python
data.head()
```

```python
experiment = ClusteringExperiment()
experiment.setup(data = data, session_id = 192, use_gpu = True)
```

```python
kmeans = experiment.create_model("kmeans")
results = experiment.assign_model(kmeans)
experiment.evaluate_model(kmeans)
```

```python
new_data = data.copy()
predictions = experiment.predict_model(kmeans, data = new_data)
predictions
```
