```python
from pycaret.time_series import *
import pandas as pd
from google.colab import files
import io
import numpy as np
```

```python
uploaded = files.upload()
data = pd.read_csv(io.BytesIO(uploaded["Metro_Interstate_Traffic_Volume.csv"]))
data.drop_duplicates(keep = False)
data["index"] = pd.to_datetime(data["date_time"])
data.set_index("index", inplace = True)
data = data.reset_index().drop_duplicates(subset='index').set_index('index')
data = data.asfreq('H')
data.head()
```

```python
data.index = pd.to_datetime(data.index).to_series()
data.index
```

```python
exp_s = setup(data, fh = 12, fold = 10, session_id = 123,
              target = "traffic_volume",
              enforce_exogenous = False,
              numeric_imputation_exogenous="nearest",
              numeric_imputation_target="nearest")
```

```python
best_model = exp_s.compare_models()
```

```python
exp_s.plot_model(best_model)
```
