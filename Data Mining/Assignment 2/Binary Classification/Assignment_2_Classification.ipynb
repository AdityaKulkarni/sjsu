```python
from google.colab import files
import pandas as pd
import io
taining_uploaded = files.upload()
train_data = pd.read_csv(io.BytesIO(taining_uploaded["train.csv"]))
train_data.head()

test_uploaded = files.upload()
test_data = pd.read_csv(io.BytesIO(test_uploaded["test.csv"]))
test_data.head()
```

```python
from pycaret.classification import *
experiment = setup(data = train_data, target = "y", session_id = 1234, use_gpu = True)
```

```python
best_model = compare_models()
```

```python
experiment.evaluate_model(best_model)
```

```python
train_prediction = experiment.predict_model(best_model)
```

```python
test_predictions = experiment.predict_model(best_model, data = test_data)
test_predictions
```
