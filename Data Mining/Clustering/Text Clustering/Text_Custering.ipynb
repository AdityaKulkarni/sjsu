```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import AgglomerativeClustering
import numpy as np
```

```python
embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
corpus = [
    "The sun shines brightly on the sandy beach.",
"Cats are known for their independence and curiosity.",
"She bought a new laptop with the latest processor.",
"Mathematics involves the study of numbers, shapes, and patterns.",
"The river flows gently through the forest.",
"Cooking at home can be a fun and rewarding experience.",
"Basketball is a popular sport played with a round ball.",
"The novel was set in a mysterious, ancient castle.",
"He enjoys playing the guitar in his free time.",
"The smartphone has a large screen and high-resolution camera.",
"Gardening requires patience and a love for plants.",
"The museum showcases artifacts from ancient civilizations.",
"She practices yoga to improve her flexibility and peace of mind.",
"The coffee shop on the corner is a great place to relax.",
"Financial planning is essential for a secure future.",
"The painting depicts a serene landscape with mountains in the background.",
"The recipe calls for fresh herbs and spices.",
"Astronomy involves the study of stars and planets.",
"The documentary focused on the effects of climate change.",
"The train travels rapidly across the countryside.",
"Learning a new language can open up many opportunities.",
"The festival features music, food, and cultural activities.",
"He is researching renewable energy sources for his project.",
"The bakery is famous for its delicious, homemade bread.",
"Wildlife conservation is important for protecting endangered species.",
    ]
```

```python
corpus_embeddings = embedder.encode(corpus)
corpus_embeddings = corpus_embeddings /  np.linalg.norm(corpus_embeddings, axis=1, keepdims=True)
corpus_embeddings
```

```python
clustering_model = AgglomerativeClustering(n_clusters=None, distance_threshold=1.5) #, affinity='cosine', linkage='average', distance_threshold=0.4)
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_
```

```python
clustered_sentences = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []

    clustered_sentences[cluster_id].append(corpus[sentence_id])
```

```python
for i, cluster in clustered_sentences.items():
    print("Cluster ", i+1)
    print(cluster)
    print("")
```
